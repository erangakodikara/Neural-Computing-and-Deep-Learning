### Define the layers & create the model object and input & output!

### https://keras.io/guides/functional_api/

#### The Keras Functional API is a way to create complex models, such as those with multiple outputs, models with shared layers, or acyclic graphs.

#### An example of a model with multiple inputs and multiple outputs:

```python
#  Your code of 
# Keras Functional model

# Keras Functional API Model
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model

# Define input layer
inputs = Input(shape=(n_features,))

# Define hidden layers
x = Dense(64, activation='relu')(inputs)
x = Dense(32, activation='relu')(x)
x = Dense(16, activation='relu')(x)
x = Dense(8, activation='relu')(x)

# Define output layer
outputs = Dense(1, activation='sigmoid')(x)

# Create the model
functional_model = Model(inputs=inputs, outputs=outputs)

# Compile the model
functional_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Print model summary
print(functional_model.summary())

# Train the model
history_func = functional_model.fit(X_train, y_train, epochs=200, batch_size=16, verbose=1, validation_split=0.2)

# Evaluate the model
loss, acc = functional_model.evaluate(X_test, y_test, verbose=1)
print('\nFunctional Model Test Accuracy: %.3f' % acc)
```

```python
# Plot training accuracy for Functional Model
history_dict_func = history_func.history
Accuracy_func = history_dict_func['accuracy']
plt.figure(num=2, figsize=(15,7))
plt.plot(Accuracy_func, 'ro', label='Functional Model Training accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Functional API Model Training Accuracy')
plt.legend()
plt.show()
```
